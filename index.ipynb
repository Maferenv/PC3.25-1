```python
import pandas as pd
link1='https://raw.githubusercontent.com/Maferenv/PC3.25-1/main/Reserves%20of%20foreign%20exchange%20and%20gold.csv'
Ex_Au=pd.read_csv(link1)
```

```python
Ex_Au
```

```python
#Renombramiento de cada columna
Nombres=['Country','Slug','Value_in_dollars', 'Date','Rank','Region']
function = {old:new for old,new in zip(Ex_Au.columns,Nombres)}
Ex_Au.rename(columns=function, inplace=True)
Ex_Au
```

```python
#Detección de datos nulos
Ex_Au.isnull()
```

```python
#Analisamos si alguna columna tiene numeros no correspondientes
Ex_Au.Region.str.contains(pat=r'\D',
                          regex=True)
Ex_Au.Slug.str.contains(pat=r'\D',
                          regex=True)
```

```python
#Detectamos la presencia de celdas con simbolos no numéricos en la columna
pd.unique(Ex_Au.Slug[~Ex_Au.Slug.str.contains(r'\d+\.\d+',regex=True)])
pd.unique(Ex_Au.Country[~Ex_Au.Country.str.contains(r'\d+\.\d+',regex=True)])
pd.unique(Ex_Au.Region[~Ex_Au.Region.str.contains(r'\d+\.\d+',regex=True)])
```

```python
#Reemplazamos todos los signos con espacios
import re
Ex_Au['Slug']=Ex_Au['Slug'].apply(lambda x:re.sub(r'[^\w\s]', ' ', x))
Ex_Au.Slug
```

```python
#Capitalizamos cada pais de la columna Slug
Ex_Au['Slug']=Ex_Au['Slug'].str.capitalize()
Ex_Au.Slug
```

```python
#Para la columna Country analisamos que no contenga simbolos ni numeros
Ex_Au.Country.str.contains(pat=r'\D',
                          regex=True)
Ex_Au.Country[Ex_Au.Country.str.contains(pat=r'[^\w\s]',regex=True)]
```

```python
#Eliminamos todas las palabras que estén contenidas dentro de paréntesis
Ex_Au['Country']=Ex_Au['Country'].str.replace(r'\([^()]*\)', '', regex=True)
Ex_Au.Country
```

```python
#Reordenamos las palabras que comienzan con la preposición the
def reorder_country_name(country):
    if ',' in country:
        parts = country.split(',')
        return f"{parts[1].strip()} {parts[0].strip()}"
    return country

Ex_Au['Country'] = Ex_Au['Country'].apply(reorder_country_name)
Ex_Au.Country
```

```python
#Analizamos si los numeros poseen simbolos no numericos y los quitamos
Ex_Au['Value_in_dollars'] = Ex_Au['Value_in_dollars'].astype(str)
Ex_Au['Value_in_dollars'] = Ex_Au['Value_in_dollars'].str.replace(r"[^\d\.\-]", "", regex=True)
```

```python
Ex_Au.Value_in_dollars=pd.to_numeric(Ex_Au.Value_in_dollars, errors='coerce')
Ex_Au.Value_in_dollars
```

```python
#Extraemos 3 ceros hacia atrás
Ex_Au['Value_in_dollars'] = Ex_Au['Value_in_dollars'].astype(str)
Ex_Au['Value_in_dollars'] = Ex_Au['Value_in_dollars'].str.replace('0', '', 3)
Ex_Au.Value_in_dollars
```

```python
#Renombramos Value in dollars
Ex_Au.rename(columns={'Value_in_dollars':'Value_in_dollars_x10^3'}, inplace=True)
Ex_Au
```

```python
#Renombramos la tabla anterior
Resv_fng=Ex_Au
Resv_fng
```

# 2.   Combinación de los dataframes
Lo que realizaremos es combinar cada uno de los dataframes sin cambiar los elementos más importantes, mientras se eliminan los datos con menor relevancia


```python
#Elimino las columnas Slug, Date y Rank de cada dataframe
#creo una tupla con todos los dataframes
dfs = (Resv_fng,En_coms,Elec_igc,Edu_exp)
for df in dfs:
  drop_columns=['Slug', 'Date', 'Rank']
  columns_exs = [col for col in drop_columns if col in df.columns]
  if columns_exs:
    df.drop(columns_exs, axis=1, inplace=True)
```

```python
for df in dfs:
    # Ordenar el dataframe por la columna Country
    if 'Country' in df.columns:
        df.sort_values(by='Country', inplace=True)
        # Resetear el índice después de ordenar si es necesario
        df.reset_index(drop=True, inplace=True)
```

# 3. Media, mínimo y máximo de la tabla principal

```python
#Determinamos la media de cada una de las 4 variables numéricas existentes
Est_paises.info()
```

```python
variabls=['Value_in_dollars_x10^3','IGC_in_10^3KW','Btu/person_(x10^3)','Percentage of GDP (%)']
#Nos aseguramos de que las nuevas columnas sean numéricas
for col in variabls:
  Est_paises[col]=pd.to_numeric(Est_paises[col], errors='coerce')
prom='mean'
Est_paises.groupby('Region')[variabls].agg(prom)
```

```python
nueva_funcion=['mean','min','max']
Est_paises=Est_paises.groupby('Region')[variabls].agg(nueva_funcion)
```
